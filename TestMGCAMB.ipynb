```python
dfc_1 = dfc.loc[dfc['k'] == k1]
dfm_1 = dfm.loc[dfm['k'] == k1]
dfm2_1 = dfm2.loc[dfm2['k'] == k1]

dfc_2 = dfc.loc[dfc['k'] == k2]
dfm_2 = dfm.loc[dfm['k'] == k2]
dfm2_2 = dfm2.loc[dfm2['k'] == k2]

dfc_3 = dfc.loc[dfc['k'] == k3]
dfm_3 = dfm.loc[dfm['k'] == k3]
dfm2_3 = dfm2.loc[dfm2['k'] == k3]

dfc_4 = dfc.loc[dfc['k'] == k4]
dfm_4 = dfm.loc[dfm['k'] == k4]
dfm2_4 = dfm2.loc[dfm2['k'] == k4]
```

```python
# Defining the plotting functions
def plot(quantity):
    plt.semilogx(dfc_3['a'].values,  dfc_3[quantity].values,  color = 'black', linestyle = '--', label= 'CAMB')
    plt.semilogx(dfm_3['a'].values,  dfm_3[quantity].values,  color = 'C0', label= r'MGCAMB, $a_{\rm GR} = 10^{-2}$')
    plt.semilogx(dfm2_3['a'].values, dfm2_3[quantity].values, color = 'C1', label= r'MGCAMB, $a_{\rm GR} = 10^{-3}$')

    
    plt.xlabel(r'$a$')
    plt.legend(loc='lower right')
    plt.ylabel(quantity)
    plt.xlim(5e-4,1.1)
    plt.title(quantity)
    plt.show()
```

```python
def plot_diff(quantity):
    #plt.semilogx(dfc_2['a'].values,  dfc_2[quantity].values,  color = 'black', linestyle = '--', label= 'CAMB')
    plt.semilogx(dfm_2['a'].values,  (dfm_2[quantity].values-dfc_2[quantity].values)/dfc_2[quantity].values,  color = 'C0', label= r'MGCAMB, $a_{\rm GR} = 10^{-2}$')
    plt.semilogx(dfm2_2['a'].values, (dfm2_2[quantity].values-dfc_2[quantity].values)/dfc_2[quantity].values, color = 'C1', label= r'MGCAMB, $a_{\rm GR} = 10^{-3}$')

    
    plt.xlabel(r'$a$')
    plt.legend(loc='lower left')
    plt.ylabel(quantity)
    plt.xlim(5e-4,1.1)
    plt.ylim(-0.01,0.01)
    plt.title(quantity)
    plt.show()
```

```python
plot('pidot_sum')
```

```python
plot_diff('pidot_sum')
```

```python
plot('dgpi')
```

```python
plot_diff('dgpi')
```

## Radiation Streaming Approximation tests

```python
gr = np.loadtxt('./testGR_scalCls.dat')
mg1 = np.loadtxt('./testMGmnu_yesRSA_scalCls.dat')
mg2 = np.loadtxt('./testMGmnu_yesRSA2_scalCls.dat')

plt.title('nu, RSA on')
plt.semilogx(mg1[:,0], (mg1[:,1]-gr[:,1])/gr[:,1], label = 'GRtrans = 0.01')
plt.semilogx(mg2[:,0], (mg2[:,1]-gr[:,1])/gr[:,1], label = 'GRtrans = 0.001')
plt.xlabel(r'$\ell$')
plt.ylabel(r'$\Delta C_{\ell}$')
plt.legend(loc = 'upper right')
plt.show()
```
